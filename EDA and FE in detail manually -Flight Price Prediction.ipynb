## Feature Engineering process

#### 1. Date_of_Journey
###### first we will derive out some features from the "Date_of_Journey" feature, which is an object type data

```python
final_df['Date_of_Journey'].str.split('/')  ###this str.split(), split a string into a list of strings after breaking the given string by the specified separator.
```

```python
final_df['Date_of_Journey'].str.split('/').str[0]  ###it will extract all the dates that are present in all teh lists at 0th index
```

```python
final_df['date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['date']=final_df['date'].astype(int)
final_df['month']=final_df['month'].astype(int)
final_df['year']=final_df['year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
final_df.head()
```

#### 2. Arrival_time

```python
final_df.head(10)
```

```python
##now note, in this data, the date is not required, it is redundant, as for that there is already another column, so we have make this data available with only time
```

```python
final_df['Arrival_Time'].str.split(' ')
```

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head(10)
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis =1, inplace=True)
```

```python
final_df.head(5)
```

#### 3. Departure time

```python
final_df['Dept_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min']=final_df['Dep_Time'].str.split(':').str[1]
final_df['Dept_hour']=final_df['Dept_hour'].astype(int)
final_df['Dept_min']=final_df['Dept_min'].astype(int)
final_df.drop('Dep_Time', axis =1, inplace=True)
```
